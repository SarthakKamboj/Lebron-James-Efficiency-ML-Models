```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import os
```

<h2>Collecting Statistical Averages from 2003 to 2020</h2>

```python
lbj_avg_res = get("https://www.basketball-reference.com/players/j/jamesle01.html")
lbj_soup = BeautifulSoup(lbj_avg_res.text,"html.parser")
```

```python
avg_df_columns = []
avg_table = lbj_soup.find("table",{"id":"per_game"})
```

```python
cols = avg_table.find("thead").find("tr").find_all("th")
for col in cols:
    avg_df_columns.append(col.text)
avg_df_columns
```

```python
avg_df = pd.DataFrame(columns=avg_df_columns)
avg_df.head()
```

```python
yearly_avgs = avg_table.find("tbody").find_all("tr")
idx = 0
for yearly_avg in yearly_avgs:
    stats = []
    season = yearly_avg.find("th").text
    stats.append(season)
    stat_tds = yearly_avg.find_all("td")
    for stat_td in stat_tds:
        stats.append(stat_td.text)
    avg_df.loc[idx] = stats
    idx += 1
avg_df.head(20)
```

```python
str_cols = ["Season","Tm","Lg","Pos"]
```

```python
num_cols = avg_df.columns.drop(str_cols)
num_cols
```

```python
for num_col in num_cols:
    avg_df[num_col] = avg_df[num_col].apply(lambda x: float(x))
avg_df.head()
```

```python
avg_df.to_csv(os.path.join("data","lebron_james_avg_stats.csv"))
```

<h2>Collecting Game Statistics of Games from 2003 to 2020</h2>

```python
def transform(n):
    try:
        return float(n)
    except:
        return np.nan

def transform_df(df):
    str_cols = ["Date","Tm","game_location","Opp","game_result","MP","Age"]
    num_cols = df.columns.drop(str_cols)
    for num_col in num_cols:
        df[num_col] = df[num_col].apply(lambda x: transform(x))
```

```python
base_url = "https://www.basketball-reference.com/players/j/jamesle01/gamelog"
start_year = 2003
end_year = 2020
for year in range(start_year+1,end_year+1):
    url = f"{base_url}/{year}/"
    year_res = get(url)
    year_soup  = BeautifulSoup(year_res.text,"html.parser")
    year_table = year_soup.find("table",{"id":"pgl_basic"})
    
#     get columns names
    year_head = year_table.find("thead").find("tr")
    col_names = []
    cols = year_head.find_all("th")
    for col in cols:
        if col.text == "Rk":
            continue
        title = col.text
        if not str(ascii(col.text)).replace("'","")==col.text:
            title = col["data-stat"]
        col_names.append(title)
    year_df = pd.DataFrame(columns=col_names)
    
#     get data
    year_body = year_table.find("tbody")
    game_rows = year_body.find_all("tr")
    
    idx = 0
    
    for game_row in game_rows:
        try:
            game_id = game_row["id"]
            data = []
            cols = game_row.find_all("td")
            for col in cols:
                data.append(col.text)
            year_df.loc[idx] = data
            idx += 1
        except:
            pass
        
    transform_df(year_df)
    
    year_df.to_csv(os.path.join("data",f"lbj_data_{year}.csv"))
    
```
